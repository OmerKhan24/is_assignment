## 3. CIA Triad Implementation <a name="cia"></a>

### ğŸ”’ Confidentiality (Privacy Protection)

**Implementation:**
1. **Data Anonymization**: Patient names converted to ANON_XXXX format
2. **Contact Masking**: Phone numbers masked (XXX-XXX-XXXX)
3. **Encryption**: Fernet symmetric encryption for reversible data protection
4. **Hashing**: SHA-256 for password storage
5. **Role-Based Access Control (RBAC)**: 
   - Admin: Full access to raw and anonymized data
   - Doctor: Access to anonymized data only
   - Receptionist: Can add/edit but cannot view sensitive data

### âœ… Integrity (Data Accuracy & Accountability)

**Implementation:**
1. **Activity Logging**: All user actions recorded with timestamps
2. **Data Validation**: Input validation before database insertion
3. **SQL Injection Prevention**: Input sanitization
4. **Database Constraints**: Primary keys, foreign keys, and CHECK constraints
5. **Audit Trail**: Complete history of who did what and when

### ğŸŒ Availability (System Access & Reliability)

**Implementation:**
1. **Error Handling**: Try-except blocks for graceful error recovery
2. **Data Backup**: CSV/JSON export functionality
3. **Session Management**: Persistent user sessions
4. **System Monitoring**: Uptime tracking and display
5. **Database Connection Pooling**: Efficient connection management

## 5. Privacy & Anonymization <a name="privacy"></a>

Demonstrating data anonymization and encryption features.

````python
# Initialize Privacy Manager
privacy_mgr = PrivacyManager()
print("âœ… Privacy Manager initialized!")

# Demo: Data Masking
sample_data = {
    'patient_id': 1,
    'name': 'John Doe',
    'contact': '123-456-7890',
    'diagnosis': 'Hypertension and Diabetes'
}

print("\nğŸ“‹ Original Patient Data:")
for key, value in sample_data.items():
    print(f"  {key}: {value}")

# Apply anonymization
anonymized = privacy_mgr.anonymize_patient_data(sample_data, sample_data['patient_id'])

print("\nğŸ”’ Anonymized Patient Data:")
for key, value in anonymized.items():
    print(f"  {key}: {value}")
````

````python
# Demo: Encryption/Decryption (Bonus Feature)
sensitive_info = "Patient has a history of heart disease"

print(f"ğŸ“ Original Text: {sensitive_info}")

# Encrypt
encrypted = privacy_mgr.encrypt_data(sensitive_info)
print(f"\nğŸ” Encrypted: {encrypted[:50]}...")

# Decrypt
decrypted = privacy_mgr.decrypt_data(encrypted)
print(f"\nğŸ”“ Decrypted: {decrypted}")

print("\nâœ… Encryption is reversible - useful for authorized access!")
````

## 6. Role-Based Access Control Demo <a name="rbac"></a>

````python
# Add sample patient
patient_id = db.add_patient(
    name="Jane Smith",
    contact="555-123-4567",
    diagnosis="Diabetes Type 2"
)

print(f"âœ… Added patient with ID: {patient_id}")

# Get patient data
patients = db.get_patients()
if patients:
    patient = patients[0]
    
    print("\nğŸ‘¨â€ğŸ’¼ ADMIN VIEW (Full Access):")
    admin_view = AccessControl.filter_patient_data(patient, 'admin')
    for key, value in admin_view.items():
        print(f"  {key}: {value}")
    
    print("\nğŸ‘¨â€âš•ï¸ DOCTOR VIEW (Anonymized):")
    doctor_view = AccessControl.filter_patient_data(patient, 'doctor')
    for key, value in doctor_view.items():
        print(f"  {key}: {value}")
    
    print("\nğŸ‘¨â€ğŸ’¼ RECEPTIONIST VIEW (Confidential):")
    receptionist_view = AccessControl.filter_patient_data(patient, 'receptionist')
    for key, value in receptionist_view.items():
        print(f"  {key}: {value}")
````

## 10. Testing & Verification <a name="testing"></a>

````python
# Test 1: Authentication
print("ğŸ§ª Test 1: User Authentication")
test_user = db.verify_user("admin", "admin123")
if test_user:
    print(f"  âœ… Login successful: {test_user['username']} ({test_user['role']})")
else:
    print("  âŒ Login failed")

# Test 2: Invalid login
invalid_user = db.verify_user("admin", "wrongpassword")
if not invalid_user:
    print("  âœ… Invalid credentials correctly rejected")
else:
    print("  âŒ Security breach: invalid login accepted")
````

````python
# Test 2: Role-Based Access Control
print("\nğŸ§ª Test 2: RBAC Permissions")

roles = ['admin', 'doctor', 'receptionist']
permissions = ['view_raw_data', 'view_anonymized_data', 'add_patient', 'view_logs']

for role in roles:
    print(f"\n  {role.upper()}:")
    for perm in permissions:
        has_perm = AccessControl.has_permission(role, perm)
        status = "âœ…" if has_perm else "âŒ"
        print(f"    {status} {perm}")
````

````python
# Test 3: Data Integrity (Logging)
print("\nğŸ§ª Test 3: Activity Logging (Integrity)")

initial_log_count = len(db.get_logs())
print(f"  Initial log count: {initial_log_count}")

# Perform action
db.add_log(1, "admin", "admin", "TEST_ACTION", "Testing logging system")

final_log_count = len(db.get_logs())
print(f"  Final log count: {final_log_count}")

if final_log_count > initial_log_count:
    print("  âœ… Logging system working correctly")
else:
    print("  âŒ Logging system failed")
````

````python
# Test 4: Data Export (Availability)
print("\nğŸ§ª Test 4: Data Backup/Export (Availability)")

patients = db.get_patients()
if patients:
    df_export = pd.DataFrame(patients)
    
    # Export to CSV
    csv_filename = f"backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
    df_export.to_csv(csv_filename, index=False)
    print(f"  âœ… Data exported to {csv_filename}")
    print(f"  ğŸ“Š Exported {len(patients)} records")
else:
    print("  âš ï¸ No data to export")
````
